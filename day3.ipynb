```python
# sum of n numbers:
n = int(input("Enter a Number:"))
sum = 0
for i in range(1,n+1):
  sum = sum + i
  print("Sum of n numbers:",sum)
```

```python
num = int(input("Enter number: "))
rev = 0

while num > 0:
    rev = rev * 10 + num % 10
    num = num // 10

print("Reverse:", rev)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sales = [100, 150, 200, 250, 300]
data = {
    "Name": ["A", "B", "C", "D"],
    "Salary": [30000, 40000, 35000, 45000]
}

df = pd.DataFrame(data)

sns.barplot(x="Name", y="Salary", data=df)
plt.title("Salary Comparison")
plt.show()


plt.plot(sales)
plt.title("Sales Growth")
plt.xlabel("Time")
plt.ylabel("Sales")
plt.show()
```
